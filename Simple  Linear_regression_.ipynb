```python
def coef_cal(x,y,data):
    a=data[x]-data[x].mean()
    b=data[y]-data[y].mean()
    ab=a*b
    aa=a*a
    beta1=sum(ab)/sum(aa)
    beta0=(data[y].mean())-(beta1*data[x].mean())
    predicted=beta0+(beta1*data[x])
    error=np.square(data[y]-predicted)
    SST=np.square(data[y]-data[y].mean())
    SSR=sum(SST)-sum(error)
    r_square=SSR/sum(SST)
    MSR=SSR/1
    MSE=(sum(error)/(len(data)-2))
    F=MSR/MSE
    print("beta1 value are",beta1)
    print("beta0 value are",beta0)
    print("SSE values are",sum(error))
    print("SST values are",sum(SST))
    print("SSR values are",SSR)
    print("rsquare value",r_square)
    print("F-stat value",F)
```

```python
coef_cal(x="Promotion_Budget",y="Passengers",data=air)
```
